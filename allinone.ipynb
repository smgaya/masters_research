```python
#import libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
#from sklearn.svm import SVM
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import svm, datasets

#import data file 

df = pd.read_csv("/home/mgaya/venv/mgaya/data/og3.csv")  

#show row data 
df.head()

dataset = pd.DataFrame(df)
dataset.columns
target = dataset["Class"]
features = dataset.drop(["datetime","Class"], axis=1)

#split 
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 10)


#training svm linear kernel  
svm_model = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(X_train, y_train)
svm_model.fit(X_train, y_train)
svm_prediction = svm_model.predict(X_test)


# training a DescisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train)
dtree_predictions = dtree_model.predict(X_test)

# training a KNN classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)

# training a Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)


#-=================================================
# retrieve the accuracy svm
svm_accuracy = svm_model.score(X_test, y_test)
print( svm_accuracy)
# creating a confusion matrix
cm_lin = confusion_matrix(y_test, svm_prediction)
#print(cm_lin)


# creating a confusion matrix
cm_dt = confusion_matrix(y_test, dtree_predictions)
#print(cm_dt)



# accuracy on X_test knn
knn_accuracy = knn.score(X_test, y_test)
print (knn_accuracy)

# creating a confusion matrix
knn_predictions = knn.predict(X_test)
cm_knn = confusion_matrix(y_test, knn_predictions)


# accuracy on X_test naive bayers
nb_accuracy = gnb.score(X_test, y_test)
print (nb_accuracy)

# creating a confusion matrix
cm_nb = confusion_matrix(y_test, gnb_predictions)
```
